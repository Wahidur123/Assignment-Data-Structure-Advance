Q5.Write a code to find all occurrences of a given substring within another string?

```python
def find_all_occurrences(main_string, substring):
    # Initialize an empty list to store the indices of occurrences
    indices = []

    # Start searching from the beginning of the string
    start = 0

    # Loop to find each occurrence of the substring
    while True:
        # Find the index of the substring starting from 'start'
        index = main_string.find(substring, start)

        # If find() returns -1, the substring is no longer found, break the loop
        if index == -1:
            break

        # Add the index to the list
        indices.append(index)

        # Move start to the next position after the found substring
        start = index + 1

    return indices

# Example usage:
main_string = input("Enter the main string: ")
substring = input("Enter the substring: ")

# Get all occurrences of the substring
occurrences = find_all_occurrences(main_string, substring)

# Output the result
if occurrences:
    print(f"The substring '{substring}' is found at indices: {occurrences}")
else:
    print(f"The substring '{substring}' is not found in the string.")
```

Q6.Write a code to perform basic string compression using  the counts of repeated characters.

```python
def compress_string(s):
    # Edge case: if the string is empty, return it as is
    if not s:
        return s

    # Initialize the compressed string and a count variable
    compressed = []
    count = 1

    # Iterate through the string, starting from the second character
    for i in range(1, len(s)):
        # If the current character is the same as the previous one, increment the count
        if s[i] == s[i - 1]:
            count += 1
        else:
            # Otherwise, append the previous character and its count to the result
            compressed.append(s[i - 1] + str(count))
            count = 1  # Reset the count for the new character

    # Don't forget to append the last character and its count
    compressed.append(s[-1] + str(count))

    # Join the list into a string
    compressed_string = ''.join(compressed)

    # If the compressed string is not shorter, return the original string
    return compressed_string if len(compressed_string) < len(s) else s

# Example usage:
input_string = "aaabbbcccaaa"
result = compress_string(input_string)
print(result)  # Output: "a3b3c3a3"
```

Q7.write a code to determine if a string has all unique characters.

```python
def has_unique_characters(s):
    # Create an empty set to track characters we've seen
    seen = set()

    # Iterate through each character in the string
    for char in s:
        # If the character is already in the set, return False
        if char in seen:
            return False
        # Otherwise, add the character to the set
        seen.add(char)

    # If we finish the loop without finding duplicates, return True
    return True

# Example usage:
input_string = "abcdefg"
result = has_unique_characters(input_string)
print(result)  # Output: True

input_string = "hello"
result = has_unique_characters(input_string)
print(result)  # Output: False
```

Q8.write a code to convert a given string to uppercase  or lowercase.

```python
def convert_string_case(s, to_upper=True):
    """
    Convert the string to either uppercase or lowercase based on the `to_upper` flag.

    :param s: The input string to convert.
    :param to_upper: If True, convert to uppercase. If False, convert to lowercase.
    :return: The string converted to the desired case.
    """
    if to_upper:
        return s.upper()  # Convert to uppercase
    else:
        return s.lower()  # Convert to lowercase

# Example usage:
input_string = "Hello, World!"

# Convert to uppercase
result_upper = convert_string_case(input_string, to_upper=True)
print("Uppercase:", result_upper)  # Output: "HELLO, WORLD!"

# Convert to lowercase
result_lower = convert_string_case(input_string, to_upper=False)
print("Lowercase:", result_lower)  # Output: "hello, world!"
```

Q12.Implement a code to find the second largest number in a given list of integers.

```python
def second_largest(numbers):
    # Remove duplicates by converting the list to a set and then back to a sorted list
    unique_numbers = sorted(set(numbers), reverse=True)

    # Check if the list has at least two distinct numbers
    if len(unique_numbers) < 2:
        return None  # Return None if there is no second largest number

    # Return the second element from the sorted list
    return unique_numbers[1]

def main():
    # Example list of integers
    numbers = [10, 20, 4, 45, 99, 45]

    # Find the second largest number
    result = second_largest(numbers)

    if result is not None:
        print(f"The second largest number is: {result}")
    else:
        print("There is no second largest number.")

# Run the main function
if __name__ == "__main__":
    main()
```

Q14.Write a code to reverse a list in-place without using any built-in reverse functions.

```python
def reverse_list_in_place(lst):
    # Use two pointers to swap elements from both ends
    left = 0
    right = len(lst) - 1

    while left < right:
        # Swap the elements
        lst[left], lst[right] = lst[right], lst[left]

        # Move the pointers towards the center
        left += 1
        right -= 1

    return lst

def main():
    # Example list
    my_list = [1, 2, 3, 4, 5]

    # Reverse the list in place
    reversed_list = reverse_list_in_place(my_list)

    # Print the reversed list
    print("Reversed list:", reversed_list)

# Run the main function
if __name__ == "__main__":
    main()
```

Q15.Implement a code to find and remove duplicates from a list while preserving the original order of elements.

```python
def remove_duplicates(lst):
    seen = set()  # A set to track seen elements
    result = []   # List to store the result without duplicates

    for item in lst:
        if item not in seen:
            result.append(item)  # Add the item to the result list if it's not a duplicate
            seen.add(item)       # Mark this item as seen

    return result

def main():
    # Example list with duplicates
    my_list = [1, 2, 2, 3, 4, 4, 5, 1, 6, 3]

    # Remove duplicates while preserving the original order
    unique_list = remove_duplicates(my_list)

    # Print the list without duplicates
    print("List without duplicates:", unique_list)

# Run the main function
if __name__ == "__main__":
    main()
```

Q17.Write a code to merge two sorted lists into a single sorted list.

```python
def merge_sorted_lists(list1, list2):
    # Initialize pointers for both lists and the result list
    i, j = 0, 0
    merged_list = []

    # Loop through both lists and compare elements
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # If any elements remain in list1, add them to the merged list
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # If any elements remain in list2, add them to the merged list
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

def main():
    # Example sorted lists
    list1 = [1, 3, 5, 7]
    list2 = [2, 4, 6, 8]

    # Merge the two sorted lists
    result = merge_sorted_lists(list1, list2)

    # Print the merged list
    print("Merged list:", result)

# Run the main function
if __name__ == "__main__":
    main()
```

Q18.Implement a code to find the intersection of two given lists.

```python
def intersection_of_lists(list1, list2):
    # Convert both lists to sets to remove duplicates and perform the intersection
    set1 = set(list1)
    set2 = set(list2)

    # Find the intersection of both sets
    intersection = set1 & set2

    # Convert the result back to a list to return it
    return list(intersection)

def main():
    # Example lists
    list1 = [1, 2, 3, 4, 5]
    list2 = [4, 5, 6, 7, 8]

    # Find the intersection of the two lists
    result = intersection_of_lists(list1, list2)

    # Print the result
    print("Intersection:", result)

# Run the main function
if __name__ == "__main__":
    main()
```

Q20.Write a code to shuffle a given list randomly without using any built-in shuffle functions.

```python
import random

def shuffle_list(lst):
    # Fisher-Yates (Knuth) Shuffle
    n = len(lst)

    # Iterate over the list in reverse order
    for i in range(n - 1, 0, -1):
        # Pick a random index between 0 and i (inclusive)
        j = random.randint(0, i)

        # Swap the elements at indices i and j
        lst[i], lst[j] = lst[j], lst[i]

    return lst

def main():
    # Example list to shuffle
    my_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

    # Shuffle the list
    shuffled_list = shuffle_list(my_list)

    # Print the shuffled list
    print("Shuffled list:", shuffled_list)

# Run the main function
if __name__ == "__main__":
    main()
```

Q32.Write a code that takes two dictionaries as input and merges them into a single dictionary.If there are common keys, the value should be added together.  

```python
def merge_dicts(dict1, dict2):
    # Create a new dictionary to store the merged result
    merged_dict = dict1.copy()  # Start with a copy of dict1

    # Iterate over the second dictionary and add values to the merged dictionary
    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value  # If the key exists, add the value
        else:
            merged_dict[key] = value  # If the key doesn't exist, just add it

    return merged_dict

def main():
    # Example input for two dictionaries
    dict1 = {'apple': 10, 'banana': 5, 'cherry': 3}
    dict2 = {'banana': 7, 'cherry': 2, 'date': 4}

    # Merge the dictionaries
    result = merge_dicts(dict1, dict2)

    # Print the result
    print("Merged dictionary:", result)

# Run the main function
if __name__ == "__main__":
    main()
```

Q33.Write a code to access a value in a nested dictionary. The function should take the dictionary and a list of keys as input, and return the corresponding value. If any of the keys do not exist in the dictionary, the function should return None.

```python
def access_nested_value(d, keys):
    # Iterate over the list of keys
    for key in keys:
        # Check if the key exists in the current dictionary level
        if isinstance(d, dict) and key in d:
            d = d[key]  # Move to the next level of the dictionary
        else:
            return None  # Return None if the key is not found or if it's not a dictionary
    return d  # Return the final value found

def main():
    # Example of a nested dictionary
    nested_dict = {
        'a': {
            'b': {
                'c': 5
            },
            'd': 10
        },
        'e': 20
    }

    # Example input for keys
    keys1 = ['a', 'b', 'c']
    keys2 = ['a', 'x']  # Invalid key path

    # Access values using the keys
    value1 = access_nested_value(nested_dict, keys1)
    value2 = access_nested_value(nested_dict, keys2)

    # Print the results
    print(f"Value for keys {keys1}: {value1}")  # Output: 5
    print(f"Value for keys {keys2}: {value2}")  # Output: None

# Run the main function
if __name__ == "__main__":
    main()
```

Q34.Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order.

```python
def sort_dict_by_value(d, descending=False):
    # Sort the dictionary by values, and return a sorted dictionary
    # Use sorted() and pass a lambda function to sort by values
    sorted_items = sorted(d.items(), key=lambda item: item[1], reverse=descending)

    # Convert the sorted items back into a dictionary
    sorted_dict = dict(sorted_items)

    return sorted_dict

def main():
    # Example dictionary
    my_dict = {'apple': 10, 'banana': 5, 'cherry': 20, 'date': 15}

    # Sort the dictionary in ascending order by value
    sorted_asc = sort_dict_by_value(my_dict, descending=False)
    print("Sorted by value (ascending):", sorted_asc)

    # Sort the dictionary in descending order by value
    sorted_desc = sort_dict_by_value(my_dict, descending=True)
    print("Sorted by value (descending):", sorted_desc)

# Run the main function
if __name__ == "__main__":
    main()
```

Q35.Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.

```python
def invert_dict(d):
    # Create an empty dictionary to store the inverted key-value pairs
    inverted_dict = {}

    # Iterate over the items in the original dictionary
    for key, value in d.items():
        if value in inverted_dict:
            # If the value is already a key in the inverted dictionary, append the key to the list
            inverted_dict[value].append(key)
        else:
            # Otherwise, add a new entry with the value as the key and the key as a list
            inverted_dict[value] = [key]

    return inverted_dict

def main():
    # Example dictionary
    my_dict = {'apple': 10, 'banana': 5, 'cherry': 10, 'date': 15}

    # Invert the dictionary
    inverted = invert_dict(my_dict)

    # Print the inverted dictionary
    print("Inverted dictionary:", inverted)

# Run the main function
if __name__ == "__main__":
    main()
```
